```python
import pandas as pd
```

```python
df = pd.read_csv("../dataset/2014-08-01.csv")
df = df[['date','time','REPORT_ID','DISTANCE_IN_METERS','vehicleCount']]
df.head(5)
```

```python
#change int to str
df['time'] = df['time'].astype(str)
df['time'] = df['time'].apply(lambda x: str(x) if (len(x)==2) else '0'+x) 
df['time'] = df['time'].apply(lambda x: x+':00:00')
df['timestamp'] = df['date']+' '+df['time']
df.head(5)
```

```python
#drop date and time inplace
df['day_of_week'] = pd.to_datetime(df['timestamp']).dt.dayofweek
df.head(10)
```
